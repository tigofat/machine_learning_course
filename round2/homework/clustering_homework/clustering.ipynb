```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from importlib import reload

import clustering

clustering = reload(clustering)

from clustering import k_means, MeasureFunctions
```

```python
rand = np.random.RandomState(123)

# info about data
length = 250
cluster_count = 3
ndim = 2

cluster_X = []
cluster_y = []

for i, (x_off, y_off) in zip(range(cluster_count), [(0, 0), (15, 15), (25, 0)]):
    cluster_X.append(rand.normal((x_off, y_off), 3, (length, ndim)))
    cluster = np.zeros((length, cluster_count))
    cluster[:, i] = 1
    cluster_y.append(cluster)
    
X = np.concatenate(cluster_X)
y = np.concatenate(cluster_y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y);
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.4, 
                                                    random_state=123)
```

```python
k_means_alg = k_means(cluster_count, MeasureFunctions.distances_measure)
```

```python
k_means_alg.train(X_train)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.scatter(k_means_alg.nk[:, 0], k_means_alg.nk[:, 1], c='black', marker='+')
plt.show()
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.scatter(k_means_alg.k[0][:, 0], k_means_alg.k[0][:, 1], c='black', marker='+')
plt.show()
```
