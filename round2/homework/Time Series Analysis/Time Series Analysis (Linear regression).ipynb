```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(12)

diff = 0.3
time = 100

data = []
for i in range(time):
    y_ = data[i-1] if data else 0
    data.append(y_ + np.random.uniform(-diff, diff))
    
x = np.arange(time)
y = np.array(data)
```

```python
plt.plot(x, data)
plt.show()
```

*y = wx + b* regression model.

Using Least Squared Method.

Loss RSS.

```python
w = len(x) * (x * y).sum() - x.sum() * y.sum()
w /= len(x) * (x**2).sum() - x.sum()**2
w
```

```python
b = (y.sum() - w * x.sum()) / len(x)
b
```

```python
xs = [x.min(), x.max()]
ys = [w * x + b for x in xs]

plt.plot(x, data)
plt.plot(xs, ys)
plt.show()
```
